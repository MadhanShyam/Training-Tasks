```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.binomial(n=10, p=0.5, size=1000)
```

```python
plt.hist(data, bins=range(12), density=True, alpha=0.75, edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
mean_simulated = np.mean(data)
print(f"Simulated Mean: {mean_simulated:.2f}")
print("Expected Mean:", 10 * 0.5)
```

```python
if abs(mean_simulated - 5) < 0.1:
    print("Matches binomial expectation (n × p = 5).")
else:
    print("Deviation observed due to randomness.")
```

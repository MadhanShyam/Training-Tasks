```python
import numpy as np
import pandas as pd 

df = pd.read_csv("Housing.csv")
df
```

```python
df[["price","area"]]
```

```python
df[['price','area']].isnull().sum()
```

```python
df.dropna(subset=['price','area'],inplace=True)
```

```python
price_col="price"
area_col="area"
```

```python
df["price_per_area"]=df[price_col] / df[area_col]
```

Threading Concept starts from Here 

```python
import threading
```

```python
def numbers():
    print([1, 2, 3])
```

```python
def letters():
    print(['A','B','C'])
```

```python
thread1=threading.Thread(target=numbers)
thread2=threading.Thread(target=letters)
```

```python
thread1.start()
thread2.start()
```

```python
thread1.join()
thread2.join()
```
